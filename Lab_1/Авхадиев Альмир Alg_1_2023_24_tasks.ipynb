###  Задание 3.
Упростите выражение
$$
\frac{(u+v)^2-4uv}{u-v}+2v-1
$$

```python
x = Symbol('u')
y = Symbol('v')

expr_1 = expand((x + y) ** 2)
expr_2 = simplify(expr_1 / (x - y) + (2 * y - 1))
display(collect(expr_2, ))
```

### Задание 7

## Матрицы
### Задание 9
Составить матрицу  $A_{3\times 4}$ из чисел вида $3k+1$, $k=1,2,...$:
$$
A=
\left(
\begin{matrix}
4& 7& 10& 13\\
16& 19& 22& 25\\
28& 31& 34& 37
\end{matrix}
\right)
$$

```python
matrix = []
k = 1
for i in range(3):
  row = []
  for j in range(k, k + 4):
    row.append(3 * j + 1)
  k += 4
  matrix.append(row)

m = Matrix(matrix)
display(m)
```

### Задание 10
В матрице А задания 9 удалить вторую строку и вставить третий столбец между первым и вторым (сам третий столбец не удалять!), должно получиться
$$
A=
\left(
\begin{matrix}
4& 10& 7& 10& 13\\
28& 34& 31& 34& 37
\end{matrix}
\right)
$$

```python
m.row_del(1)
paste = m.col(2)

display(m.col_insert(1, paste))

```
